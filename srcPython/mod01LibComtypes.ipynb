## Coneccion y acceso a funciones de Excel

```python
try:
	# Connecting | coneccion
	xl = CreateObject("Excel.Application")
except (OSError, comtypes.COMError):
	print("No tiene instalada el programa(Excel).")
	sys.exit(-1)

xl.Visible = True
print (xl)
```

```python
dir(xl)
```

```python
xl.Workbooks.Add()
# help(engine)

from comtypes.gen.Excel import xlRangeValueDefault

xl.Range["A1", "C1"].Value[xlRangeValueDefault] = (10,"20",31.4)


# print (xl.Range["A1", "C1"].Value[xlRangeValueDefault](10, "20", 31.4))
print (xl.Range["A1", "C1"].Value[xlRangeValueDefault])

print (xl.Range["A1", "C1"].Value())

xl.Range["A1", "C1"].Value[:] = (3, 2, 1)
xl.Range["A1", "C1"].Value[()] = (1, 2, 3)

print (xl.Range["A1", "C1"].Value[:])
print (xl.Range["A1", "C1"].Value[()])
```

## Acceso a otras aplicaciones

Conectandome a Speech de windows...

```python
sapi = comtypes.client.CreateObject("SAPI.SpVoice")
sapi.Speak(
    """
    Hola Francisco!. En que te puedo ayudar?
    """)
print(sapi)
```

```python
# comtypes.client.GetModule(('{C866CA3A-32F7-11D2-9602-00C04F8EE628}',))
comtypes.client.GetModule("C:\\WINDOWS\\System32\\Speech\\Common\\sapi.dll")
import comtypes.gen.SpeechLib as sapi_dll

sapi = comtypes.client.CreateObject(sapi_dll.SpVoice, interface=sapi_dll.ISpeechVoice)
sapi.Speak("Hola!, Soy parte de la libreria `comtypes`, y puedo leer lo que escribais en estas lineas de parrafo.")
print(sapi)
```
